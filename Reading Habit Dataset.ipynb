#### Conclusion:

The Reading Habit Dataset analysis revealed clear links between demographic and socioeconomic characteristics and people’s reading behaviors. Higher education and income levels correlated with a greater diversity in reading habits, including the use of printed books, e-books, and audiobooks. Younger participants showed more flexibility toward digital formats, while older groups tended to favor traditional printed books and newspapers. Employment status and marital status also influenced reading frequency and preferences, indicating that lifestyle factors play a major role in shaping reading engagement.

Notably, a significant portion of the population reported reading few or no books within the past year, emphasizing ongoing challenges in promoting consistent reading habits. By combining statistical exploration and visual analysis, this study highlights how access, affordability, and personal background shape an individual’s engagement with reading. These findings can inform literacy campaigns, library initiatives, and educational policies designed to foster lifelong learning and make reading more inclusive across different communities.
